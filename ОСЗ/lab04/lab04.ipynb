```python
import cv2 as cv
from IPython.display import display, Image
import matplotlib.pyplot as plt

# Імпорт кольорового зображення
color_image = cv.imread('img/obezyana-2.jpg') 

# Перетворення кольорового зображення в RGB формат
color_image_rgb = cv.cvtColor(color_image, cv.COLOR_BGR2RGB)

# Імпорт напівтонового зображення (відтінки сірого)
gray_image = cv.imread('img/obezyana-2.jpg', cv.IMREAD_GRAYSCALE)

# Візуалізація кольорового зображення
plt.imshow(color_image_rgb)
plt.title('Кольорове зображення')
plt.axis('off')  # Вимкніть вісі x та y
plt.show()

# Візуалізація напівтонового зображення
plt.imshow(gray_image, cmap='gray')
plt.title('Напівтонове зображення')
plt.axis('off')  # Вимкніть вісі x та y
plt.show()
```

1.Що таке pip і conda та чим вони відрізняються між собою?

`pip` (Python Package Installer) - це стандартний інструмент управління пакетами для Python. Він призначений для інсталяції, оновлення та видалення пакетів Python. `pip` дозволяє легко встановлювати сторонні бібліотеки та інші компоненти, необхідні для роботи з Python. `conda` дозволяє інсталювати, оновлювати, видаляти та керувати пакетами для різних мов програмування, включаючи Python. Він має свій власний репозиторій пакетів, який включає в себе багато популярних бібліотек. Основна різниця між `pip` і `conda` полягає в тому, що `pip` призначений для управління пакетами Python і працює з PyPI, тоді як `cond`a - це більш загальний інструмент, який може використовуватися для управління пакетами різних мов програмування та має власний репозиторій пакетів. conda також надає інструменти для створення і управління віртуальними середовищами Python, що робить його зручним для ізоляції середовищ та управління залежностями.

3.Як виконати візуалізацію зображення засобами mathplotlib?

Для візуалізації зображення засобами бібліотеки Matplotlib, використовується функція `imshow`. `imshow` - це функція з бібліотеки Matplotlib, яка використовується для візуалізації зображень. Вона дозволяє відображати зображення в графічному вікні, що надається Matplotlib.

```python
import matplotlib.pyplot as plt
import cv2
image = cv2.imread('img/1.jpg') 
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.title('Зображення')
plt.axis('off') 
plt.show()
```
